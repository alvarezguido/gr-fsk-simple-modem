```python
# 100K samples/seg
# 100 samples x data bit 
# Rb in Tx is 1kbps
```

```python
import numpy as np
```

```python
data = './outputBits.txt'
```

```python
byte_list = []
with open(data, "rb") as f:
    while True:
        byte = f.read(1)
        if not byte:
            break
        byte_list.append(ord(byte))
```

```python
type(byte_list[0])
```

```python
print (byte_list)
```

```python
## constantes ##
threshold = 30
threshold_0 = 100
threshold_1 = 40
```

```python
cont = 0
values = []
aux = []

aux2=np.argwhere(np.array(byte_list)>threshold)[0][0] #devuelve primer elemento mayor al umbral


for sample in byte_list[aux2:]:
    cont = cont + 1
    aux.append(sample)
    if cont == 100:
        if np.mean(aux) > (threshold_0):
            values.append(0)
        elif np.mean(aux) > (threshold_1):
            values.append(1)
        aux = []
        cont = 0
```

```python
len(values)
```

```python
my_bytes = bytes(values)
```

```python
with open("outputMessage.txt", "wb") as binary_file:
   
    # Write bytes to file
    binary_file.write(my_bytes)
```
